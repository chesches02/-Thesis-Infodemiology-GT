```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
import os

# Function to read the correlation matrix
def read_correlation_matrix(file_path):
    df = pd.read_csv(file_path, index_col=0)
    return df

# Directory containing the CSV files
directory = '../construct_network/data/processed/d_corr_correlation_matrix'

# Get all CSV files in the directory
all_files = [f for f in os.listdir(directory) if f.endswith('.csv')]

if not all_files:
    print(f"No CSV files found in the directory: {directory}")
    print("Current working directory:", os.getcwd())
    exit(1)

# Combine all correlation matrices
dataframes = []
for f in all_files:
    file_path = os.path.join(directory, f)
    try:
        df = read_correlation_matrix(file_path)
        dataframes.append(df)
    except Exception as e:
        print(f"Error reading file {file_path}: {str(e)}")

if not dataframes:
    print("No valid dataframes to concatenate. Check the file contents and format.")
    exit(1)

# Concatenate all dataframes
combined_df = pd.concat(dataframes)

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(combined_df)

# Perform PCA
pca = PCA()
pca.fit(scaled_data)

# Plotting the explained variance
plt.figure(figsize=(12, 6))
plt.plot(range(1, len(pca.explained_variance_) + 1), pca.explained_variance_, marker='o', linestyle='--')
plt.title('Scree Plot of Principal Components in RSV Correlation Matrix')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance')
plt.xticks(range(1, len(pca.explained_variance_) + 1))
plt.grid()
plt.tight_layout()
plt.show()

# Optionally, you can also perform PCA with a specified number of components
n_components = 2
pca = PCA(n_components=n_components)
pca_result = pca.fit_transform(scaled_data)

# Create a DataFrame for PCA results
pca_df = pd.DataFrame(data=pca_result, columns=['PCA1', 'PCA2'])

# Plotting the PCA results
plt.figure(figsize=(12, 10))
plt.scatter(pca_df['PCA1'], pca_df['PCA2'], alpha=0.6)

# Add labels for each point


plt.title('PCA of MSV Correlation Matrix')
plt.xlabel('PCA Dimension 1')
plt.ylabel('PCA Dimension 2')
plt.tight_layout()
plt.show()
```
